#Visualization: Plot the actual daily returns against the predicted returns from the stacking model and individual models for each bank. This visual comparison can reveal how well each model captures trends and outliers.

```python
import plotly.express as px
```

```python
def compare_hdfc(hdfc,cmp):
    fig=px.line(hdfc,x=hdfc['Date'],y=['Daily Return','lin','svr','st','rf'],
            template='plotly_dark',
            color_discrete_sequence=['red','orange','blue','green','yellow'],
            title='{} HDFC prediction vs actual'.format(cmp)
               )
    fig.update_layout(
    xaxis_title="Date", yaxis_title="Models"
    )
    fig.show()
```

```python
compare_hdfc(hdfc,'hdfc')
```

```python
hdfc['Volatility'] = np.sqrt(np.mean((hdfc['Daily Return']-(np.mean(hdfc['Daily Return'])))**2))
```

# Bank-Specific Insights:

- Comparative Performance: Comparing the average daily returns of each bank. 

- Correlation: Calculating the correlation coefficient between the daily returns of different banks. A high positive correlation suggests the banks' returns move together, potentially due to common factors affecting the entire banking sector. A low or negative correlation might indicate the banks are responding differently to market events.

- Risk-Return relation: Analyzing the relationship between average daily return and volatility (standard deviation) for each bank. Banks with higher returns might also exhibit higher volatility, implying a greater risk for investors.

#1 comparative performance

```python
sbi_avg = np.mean(sbi['Daily Return'])
hdfc_avg = np.mean(hdfc['Daily Return'])
icici_avg = np.mean(icici['Daily Return'])
kotak_avg = np.mean(kotak['Daily Return'])
avg = [sbi_avg,hdfc_avg,icici_avg,kotak_avg]
bank = ['sbi','hdfc','icici','kotak']
```

```python
plt.bar(bank , avg)

# Add labels and title
plt.xlabel('banks')
plt.ylabel('avg return')
plt.title('avg daily return')
```
